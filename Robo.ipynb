```python
#Teste na primeira pagina
Produto = '#__next > div > main > section:nth-child(5) > div.sc-gxoAxl.khektj > div > ul > li:nth-child(1) > a > div.sc-gQSkpc.jTodsw > h2'
navegador.find_element(By.CSS_SELECTOR, Produto).text
```
